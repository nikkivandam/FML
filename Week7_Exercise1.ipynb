```python
import pandas as pd
from sklearn.cluster import KMeans #The k-means algorithm
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df_subset = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
df_subset.head()
```

```python
import seaborn as sns #this is the plotting library I'll be using 
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs
sns.scatterplot(x="PetalLengthCm", y="SepalLengthCm", style="Species", data=df)
plt.show()
```

The tree clusters are seperated within the different species. With the naked eye, it is obvious that Iris-setosa has a high cluster. 

```python
km = KMeans(n_clusters=3) #create a new k-means model with 3 clusters
X = df[['SepalLengthCm', 'PetalLengthCm']] #get the X variables from the dataframe
km = km.fit(X) #calculate the cluster centers
df['cluster'] = km.predict(X) #predict the clusters of each observation and store in the dataframe within a new variable names 'cluster'
df.head(5)
```

```python
sns.scatterplot(x="PetalLengthCm", y="SepalLengthCm", hue="cluster", data=df)
plt.show()
```

With just comparing the two scatterplots, you see that K-means performs quit well. It clustered the different kinds of species kind of similar. Only the cluser 0 has an outlier that I would be expected to be part of the cluster 2.
