```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs
import sklearn as sk

df = pd.read_csv("funda-amsterdam.csv")
df = df.dropna()
df.head()
```

```python
sns.distplot(df['price'].dropna(), kde=False, color="c") 
plt.title('Prices of housing in Amsterdam on Funda')
plt.xlabel('Price in euros')
plt.show()
```

Checking if the dataset needs adjustments 

```python
def under_20000(x):
    if(x < 20000): 
        return float('NaN')
    else: 
        return x
df['price'] = df['price'].apply(under_20000)
df = df.dropna()

sns.distplot(df['price'].dropna(), kde=False, color="c") 
plt.title('Prices of housing in Amsterdam on Funda')
plt.xlabel('Price in euros')
plt.show()
```

Excluding all prices under 20000 euros
