```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

df = pd.read_csv("class_data.csv")
df.head()
```

```python
sns.distplot(df['texts_yesterday'].dropna(), kde=False)
plt.title('Number of texts send yesterday')
plt.xlabel('Numer of texts')
plt.show()
```

```python
sns.countplot(y="gender", data=df, color="c");
```

```python
sns.violinplot(x="apps_phone", y="phones_owned", data=df)
```

```python
sns.catplot(x="linkedin_con", y="texts_yesterday", hue="gender", kind="box", data=df)
```

```python
sns.catplot(x="like_social", kind="count", data=df)
plt.title("I like social media")
```

```python
sns.distplot(df["instagram_followers"], hist=False)
```
